**Question 4.** Relationships between variables. A scatterplot illustrates a linear relationship between two variables.
```python
data = pd.read_csv(r'C:\Users\Kolsoma\DAV_05_Data_features\ENB2012_data.csv', sep=',')


a4_dims = (18,12)
fig, ax = pyplot.subplots(figsize=a4_dims)

cmap = sns.cubehelix_palette(dark=.4,light=.8, as_cmap=True)
ax = sns.scatterplot(x="Y1", y="Y2",       #setting x and y values
                     hue="X4",             #filtering buy color using roof area
                     palette="Set2",
                     data=data,s=70)
ax.set_title("Scatter plot showing Heating Load vs Cooling Load")   #plot title
pyplot.xlabel('Heating Load')           #xlabel
pyplot.ylabel('Cooling Load')         #ylabel

'''The scatter plot tries to show the relationship between the heating load and the cooling load . We can deduce a linear 
relationship between them. Also when you add a third variable to the plot such as Roof Area. we observe that the an increase
in Roof Area results in a decrease in cooling and heating load respectively'''
```
